```python
import random

# with open("kickboxingmoves.txt","r") as infile:
#     raw_input = infile.readlines()
#     kickboxingmoves = [datum.strip('\n') for datum in raw_input]

# with open("cardiointensity.txt","r") as infile:
#     raw_input = infile.readlines()
#     cardiointensity = [datum.strip('\n') for datum in raw_input]

# with open("core.txt","r") as infile:
#     raw_input = infile.readlines()
#     core = [datum.strip('\n') for datum in raw_input]

class Workout:
    """Parent class that takes input from user on time and intensity of
    kickboxing workout."""

    def __init__(self, time):
        self.time = time
        # self.intensity = intensity

        if time not in [30, 45, 90]:
            raise Exception("Time must be either 30, 45, or 60")

        # if intensity not in [low, medium, high]:
        #     raise Exception("Intensity must be low, medium or high")

class Warmup:
    """Builds a random warm up for the workout based on the timing input by
    the user."""

    with open("warmup.txt","r") as infile:
        raw_input = infile.readlines()
        warmup = [datum.strip('\n') for datum in raw_input]

    with open("shadowboxing.txt","r") as infile:
        raw_input = infile.readlines()
        shadowboxing = [datum.strip('\n') for datum in raw_input]

    def __init__(self, time):
        self.time = time
        # self.intensity = intensity
        
    def cardiowarmup(self):
        if time == 30 or 45:
            c_keys = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec"]
        if time == 60:
            c_keys = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", 
                      "30 sec", "30 sec", "30 sec", "30 sec"]
        c_values = []
        [c_values.append(random.choice(warmup)) for key in c_keys]
        c_warmupdict = dict(zip(c_keys, c_values))
        return c_warmupdict
        
    def shadowwarmup(self):
        if time == 30 or 45:
            s_keys = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec"]
        if time == 60:
            s_keys = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", 
                      "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec"]
        s_values = []
        [s_values.append(random.choice(shadowboxing)) for key in s_keys]
        s_warmupdict = dict(zip(s_keys, s_values))
   
    def __str__(self):
        return (str(c_warmupdict) + str(s_warmupdict))
```

```python
# JUST FUNCTIONS - START HERE

import random 
  
def cardiowarmup(time):
    
    with open("warmup.txt","r") as infile:
        raw_input = infile.readlines()
        warmup = [datum.strip('\n') for datum in raw_input]
    
    if time == 30 or 45:
        c_vals = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec"]
    if time == 60:
        c_vals = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", 
                      "30 sec", "30 sec", "30 sec", "30 sec"]
    c_keys = [random.sample(warmup, len(c_vals))]
    flattened_keys = [val for sublist in c_keys for val in sublist]
    return dict(zip(flattened_keys, c_vals))
    
def shadowwarmup(time):
    
    with open("shadowboxing.txt","r") as infile:
        raw_input = infile.readlines()
        shadowboxing = [datum.strip('\n') for datum in raw_input]
    
    if time == 30 or 45:
        s_vals = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec"]
    if time == 60:
        s_vals = ["30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", 
                  "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec", "30 sec"]
    s_keys = [random.sample(shadowboxing, len(s_vals))]
    flattened_keys = [val for sublist in s_keys for val in sublist]
    return dict(zip(flattened_keys, s_vals))

def round_structure(time):
    
    with open("kickboxingmoves.txt","r") as infile:
        raw_input = infile.readlines()
        kickboxingmoves = [datum.strip('\n') for datum in raw_input]
        
    with open("cardiointensity.txt","r") as infile:
        raw_input = infile.readlines()
        cardiointensity = [datum.strip('\n') for datum in raw_input]
        
    if time == 30:
        num_rounds = 4
    if time == 45 or 60:
        num_rounds = 8
        
    combo_time = ("1 min 10 sec")
    combo_moves = [random.sample(kickboxingmoves, 5) for _ in range(num_rounds)] 
    c_intensity_time = ("20 sec")
    c_intensity_moves = [random.sample(cardiointensity, 1) for _ in range(num_rounds)]  
    
    for rnd in range(num_rounds):
        print ("Round", rnd+1, '\n', combo_time, '\n', combo_moves[rnd], '\n',
              c_intensity_time, '\n', c_intensity_moves[rnd], '\n', "switch feet to southpaw stance",
              '\n', combo_time, '\n', combo_moves[rnd])

    
```
