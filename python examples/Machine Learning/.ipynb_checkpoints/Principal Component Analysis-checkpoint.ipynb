## Project 3: Poisonous mushrooms

In this project, you'll investigate properties of mushrooms. This classic dataset contains over 8000 observations, where each mushroom is described by a variety of features like color, odor, etc., and the target variable is an indicator for whether the mushroom is poisonous. Since all the observations are categorical, I've binarized the feature space. Look at the feature_names below to see all 126 binary names.

You'll start by running PCA to reduce the dimensionality from 126 down to 2 so that you can easily visualize the data. In general, PCA is very useful for visualization (though sklearn.manifold.tsne is known to produce better visualizations). Recall that PCA is a linear transformation. The 1st projected dimension is the linear combination of all 126 original features that captures as much of the variance in the data as possible. The 2nd projected dimension is the linear combination of all 126 original features that captures as much of the remaining variance as possible. The idea of dense low dimensional representations is crucial to machine learning!

Once you've projected the data to 2 dimensions, you'll experiment with clustering using KMeans and density estimation with Gaussian Mixture Models. Finally, you'll train a classifier by fitting a GMM for the positive class and a GMM for the negative class, and perform inference by comparing the probabilities output by each model.

```python
%matplotlib inline

import pandas
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
# changed below from GMM to GaussianMixture
from sklearn.mixture import GaussianMixture
from matplotlib.colors import LogNorm
from scipy.spatial.distance import cdist
```

Load feature names.

```python
feature_names = []
with open('data/mushroom.map') as fmap:
    for line in fmap:
        [index, name, junk] = line.split()
        feature_names.append(name)

print ('Loaded feature names:', len(feature_names))
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []

with open('data/mushroom.data') as fdata:
    for line in fdata:
        items = line.split()
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print (train_data.shape, test_data.shape)
```

(1) Run Principal Components Analysis on the data. Show what fraction of the total variance in the training data is explained by the first k principal components, for k in [1, 50].

```python
def P1():
### STUDENT START ###
    # Define Principal Components Analysis
    pca = PCA()
    
    # Find the principal components for the training data
    pca.fit(train_data)
    
    # Show what fraction of the total variance in the training data is explained by the first
    # k principal components, for k in [1,50]
    k = 50
    partial_variance = pca.explained_variance_ratio_[:k].cumsum()
    
    # print results
    for k in range(k):
        print ('Component {0} explains {1:.1f}% of the total variance'.\
                   format(k+1, 100*partial_variance[k]))                                     
                                     
### STUDENT END ###

P1()
```

(2) PCA can be very useful for visualizing data. Project the training data down to 2 dimensions and plot it. Show the positive (poisonous) cases in blue and the negative (non-poisonous) in red. Here's a reference for plotting: http://matplotlib.org/users/pyplot_tutorial.html

```python
def P2():

### STUDENT START ###
    # Use n_components to pare down to the first 2 components
    pca2 = PCA(n_components = 2)
    
    # make 2-D
    train_pca = pca2.fit_transform(train_data)
    
    # Plot
    plt.figure()
    
    # blue or red depending on the train_label (1=poisonous and blue, 0=not and red)
    colors = np.where(train_labels == 1, 'blue', 'red')
    
    # use semi-transparent to show overlap
    plt.scatter(train_pca[:,0], train_pca[:,1], c = colors, alpha = 0.25, s = 50)
    plt.xlabel("Principal Component #1")
    plt.ylabel("Principal Component #2")
    plt.title("Poisonous Mushrooms in Blue, Non-Poisonous in Red")


### STUDENT END ###

P2()
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
def P3():
### STUDENT START ###
    
    # Create pca as in Problem 2, using only 2 principal components
    pca2 = PCA(n_components = 2)
    
    # Transform the coordinates into this new 2-D space
    train_pca = pca2.fit_transform(train_data)
   
    # function that plots 1-n k-means clusters for a 2 dimensional dataset
    def plotclusters(data_2D, N):
        
        # 3 subplots per row
        plt.figure(figsize = (5*min(N,3), 5*int(np.ceil(N/3.))))
        
        for n in range(1, N+1):
            # Set number of clusters and initialize them
            kmeans = KMeans(n_clusters = n, init = 'k-means++', max_iter=1000)
            
            # Fit to the data
            clusters = kmeans.fit(data_2D)
            
            # use color palette so each cluster has a different color
            colors = plt.cm.Set1(np.linspace(0, 1, n))
            
            # each color is represented by 4 coordinates
            clustercolors = np.empty(4*data_2D.shape[0]).reshape(data_2D.shape[0], 4)
            
            for i in np.arange(0, len(clusters.labels_)):
                color_index = clusters.labels_[i]
                
                # Each color is indexed by the number of the cluster
                clustercolors[i,:] = colors[color_index]
                
            # create the subplots, each element gets the color of the cluster to which 
            # it belongs
            ax = plt.subplot(int(np.ceil(N/3.)), int(np.where(N/3>=1,3,n)), n)
           
            # Plot the observations of each cluster in its corresponding color
            plt.scatter(data_2D[:,0], data_2D[:,1], c = clustercolors, alpha = 0.33, s = 25)
            
            # plot a circle that goes through the most distant point assigned to each cluster
            for j in range(n):
                # calculate radii for each of n clusters  
                centers = kmeans.cluster_centers_
                radii = [cdist(data_2D[clusters.labels_ == j], [center]).max()
                         for j, center in enumerate(centers)]
                
                for c, r in zip(centers, radii):
                    ax.add_patch(plt.Circle(c, r, fc='none', ec='black', lw=1, alpha=0.2, zorder=1))
                               
            # Plot the centroid of each cluster
            plt.scatter(clusters.cluster_centers_[:,0], clusters.cluster_centers_[:,1],
                         marker = 'x', c = 'black', s = 25)
            # title plot, make labels, set axes
            plt.title(str(n) + " cluster(s)")
            plt.xlabel("Principal Component #1")
            plt.ylabel("Principal Component #2")
            plt.xlim([np.floor(data_2D.min()), np.ceil(data_2D.max())])
            plt.ylim([np.floor(data_2D.min()), np.ceil(data_2D.max())])        

    
    # Use the defined function using the PCA dataset
    plotclusters(train_pca, N=16)
    
### STUDENT END ###

P3()
```

(4) Fit a Gaussian Mixture Model for the positive examples in your 2d projected data. Plot the estimated density contours as shown here: http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_pdf.html#example-mixture-plot-gmm-pdf-py. Vary the number of mixture components from 1-4 and the covariance matrix type ('spherical', 'diag', 'tied', 'full').

```python
def P4():
### STUDENT START ###
    
    # Use only first 2 principal components to create 2D data
    pca2 = PCA(n_components = 2)
    # Transform the coordinates into this new 2-D space
    train_pca = pca2.fit_transform(train_data)

    # Create function to plot estimated density contours
    def plot_contours_GMM(data_2D, labels, cmpnts, cvar):

        gmm = GaussianMixture(n_components = cmpnts, covariance_type = cvar)
            
        # Fit the GM model to the positive examples
        gmm.fit(train_pca[labels == 1])
                        
        # display predicted scores by the model as a contour plot
        x = np.linspace(-3.0, 3.0)
        y = np.linspace(-3.0, 3.0)
        X, Y = np.meshgrid(x, y)
        XX = np.array([X.ravel(), Y.ravel()]).T
        Z = -gmm.score_samples(XX)
        Z = Z.reshape(X.shape)
        CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=100.0),
                 levels=np.logspace(0, 2, 10))
        CB = plt.colorbar(CS, shrink=0.8, extend='both')
        colors = np.where(labels == 1, 'blue', 'red')
        plt.scatter(train_pca[:, 0], train_pca[:, 1], c = colors, alpha = 0.25, s = 50,
                edgecolor='')
        plt.title('Negative log-likelihood predicted by a GMM with\n{0} component(s) and'
                      ' {1} covariance matrix type'.\
                      format(cmpnts, cvar))
        plt.axis('tight')
        plt.show()
    
    for i in range(1,5):
        plot_contours_GMM(train_pca, train_labels, i, 'spherical')

    for i in range(1,5):
        plot_contours_GMM(train_pca, train_labels, i, 'diag')
    
    for i in range(1,5):
        plot_contours_GMM(train_pca, train_labels, i, 'tied')

    for i in range(1,5):
        plot_contours_GMM(train_pca, train_labels, i, 'full')


### STUDENT END ###

P4()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What is the accuracy?

```python
def P5():

### STUDENT START ###
    
    # Use only first 2 principal components
    pca2 = PCA(n_components = 2)
    
    # Transform the coordinates into this new 2-D space
    # for both train and test set (for predictions)
    train_pca = pca2.fit_transform(train_data)
    test_pca = pca2.transform(test_data)
    
    # Create 2 Gaussian Mixture models, one for pos examples, one for neg
    gmm_pos = GaussianMixture(n_components = 4, covariance_type = 'full')
    gmm_neg = GaussianMixture(n_components = 4, covariance_type = 'full')
    
    # Fit each GMM (1 to positive, 1 to negative)
    gmm_pos.fit(train_pca[train_labels == 1])
    gmm_neg.fit(train_pca[train_labels == 0])
    
    # Compute the log probabilities for each model
    logprob_pos = gmm_pos.score_samples(test_pca)
    logprob_neg = gmm_neg.score_samples(test_pca)
    
    # Predict the type of mushroom after comparing probabilities
    predictions = np.where(logprob_pos >= logprob_neg, 1, 0)

    # Print the accuracies
    print ('Accuracy: {0:.2f}%'.\
           format(100*metrics.accuracy_score(test_labels, predictions)))
    
### STUDENT END ###

P5()
```

(6) Ideally, we'd like a model that gives the best accuracy with the fewest parameters. Run a series of experiments to find the model that gives the best accuracy with no more than 50 parameters. For example, with 3 PCA components and 2-component diagonal covariance GMMs, you'd have:

( (3 mean vector + 3 covariance matrix) x 2 components ) x 2 classes = 24 parameters

You should vary the number of PCA components, the number of GMM components, and the covariance type.

```python
def P6():
### STUDENT START ###
    
    # n_classes will always be 2 (pos and neg)
    # For covariance_type='full', num_parameters = (n_pca*n_gmm + n_pca (n_pca + 1)/ 2 * n_gmm) * 2
        # If num_parameters <= 50, n_pca = 1 implies n_gmm <= 5, and n_gmm = 1 implies n_pca <= 12
    
    # For covariance_type='diag', num_parameters = (n_pca*n_gmm + n_pca * n_gmm) * 2
        # If num_parameters <= 50, n_pca = 1 implies n_gmm <= 12, and n_gmm = 1 implies n_pca <= 12
    
    # For covariance_type='spherical', num_parameters = (n_pca*n_gmm + n_gmm) * 2
        # If num_parameters <= 50, n_pca = 1 implies n_gmm <= 12, and n_gmm = 1 implies n_pca <= 24    
    
    # For covariance_type='tied', num_parameters = (n_pca*n_gmm + (n_pca)^2) * 2
        # If num_parameters <= 50, n_pca = 1 implies n_gmm <= 24, and n_gmm = 1 implies n_pca <= 5
    
    # So n_pca must be <= 24 and n_gmm must be <= 24 for num_parameters to stay under 50
    
    # initialize all variables, create conditions for covariance types
    n_pca, n_gmm = np.meshgrid(np.arange(1, 25), np.arange(1, 26))
    cond_full = n_pca*(n_pca + 3)*n_gmm <= 50
    cond_diag = (n_pca*n_gmm + n_pca * n_gmm) * 2 <= 50
    cond_spherical = (n_pca*n_gmm + n_gmm) * 2 <= 50
    cond_tied = (n_pca*n_gmm + (n_pca)^2) * 2 <= 50
    
    conditions = [cond_full, cond_diag, cond_spherical, cond_tied]
    covar = ['full', 'diag', 'spherical', 'tied']
    best_accuracy = 0.0
    best_covar_type = None
    best_npca = None
    best_ngmm = None

    trials = np.empty((0,4))
    
    for i,cond in enumerate(conditions):
        # For the 4 possible types of covariance matrix
        for j,k in np.column_stack((n_pca[cond],n_gmm[cond])):
            
            # For all values of n_pca and n_gmm that satisfy the condition (num_parameters <= 50)
            # Use n_pca features (principal components)
            pca_npca = PCA(n_components = j)
            train_pca = pca_npca.fit_transform(train_data)
            test_pca = pca_npca.transform(test_data) 
            
            # Create 2 GMMs, one per class, with n_gmm mixture components
            gmm_pos = GaussianMixture(n_components = k, covariance_type = covar[i])
            gmm_neg = GaussianMixture(n_components = k, covariance_type = covar[i])
            
            # Fit each GMM to the corresponding examples
            gmm_pos.fit(train_pca[train_labels == 1])
            gmm_neg.fit(train_pca[train_labels == 0])
            
            # Compute the log probabilities for each model
            logprob_pos = gmm_pos.score_samples(test_pca)
            logprob_neg = gmm_neg.score_samples(test_pca)
            
            # Predict the type of mushroom after comparing probabilities
            predictions = np.where(logprob_pos >= logprob_neg, 1, 0)
            
            # Calculate accuracy
            acc = metrics.accuracy_score(test_labels, predictions)
            
            # update experiment with new trial
            trials = np.append(trials, np.array([[j, k, i, acc]]), axis=0)
            if acc > best_accuracy:
                best_accuracy = acc
                best_npca = j
                best_ngmm = k
                best_covar_type = covar[i]
                
    # Find best combination for accuracy using less than 50 parameters
    if best_covar_type == 'full':
        num_parameters = (best_npca*best_ngmm + best_npca (best_npca + 1)/ 2 * best_ngmm) * 2
    elif best_covar_type == 'diag':
        num_parameters = (best_npca*best_ngmm + best_npca * best_ngmm) * 2
    elif best_covar_type == 'spherical':
        num_parameters = (best_npca*best_ngmm + best_ngmm) * 2
    elif best_covar_type == 'tied':
        num_parameters = (best_npca*best_ngmm + (best_npca)^2) * 2
    
    print ('Model with best accuracy using no more than 50 parameters uses')
    print ('{0} PCA components and {1}-component {2} covariance Gaussian Mixture Models'.\
        format(best_npca, best_ngmm, best_covar_type.upper()))
    print ('({0} parameters) Accuracy: {1:.1f}%\n'.\
        format(num_parameters, 100*best_accuracy))
        
### STUDENT END ###

P6()
```
